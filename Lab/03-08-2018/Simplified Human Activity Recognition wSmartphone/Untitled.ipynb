```python
def change_Activity(cols):
    Activity = cols[0]
    
    if Activity == 'STANDING':
        return 0
    if Activity == 'SITTING' :
        return 1
    if Activity == 'LAYING':
        return 2
    if Activity == 'WALKING' :
        return 3
    if Activity == 'WALKING_DOWNSTAIRS':
        return 4
    if Activity == 'WALKING_UPSTAIRS' :
        return 5
```
